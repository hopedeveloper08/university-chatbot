# Load documents

```python
import os

base_data_dir = '../data/raw/'
file_names = os.listdir(base_data_dir)

text = ''
for file_name in file_names:
    with open(os.path.join(base_data_dir, file_name), 'r', encoding='utf-8') as f:
        text += f.read()
```

# Preprocess

```python
import sys
import os
sys.path.append(os.path.abspath('../'))
from src import preprocessor

cleaned = preprocessor(text)
```

```python
with open('../data/preprocess/dataset.txt', 'w', encoding='utf-8') as f:
    f.write(cleaned)
```
